## Exploratory Analysis Plan

## 1. What are the goals of the exploratory analysis?

Answer: The goals of exploratory analysis are detection of outliers and anomalies, testing our assumptions about the data set, determine important variables, gain better understanding of the dataset by determining its main characteristics. This will determine the feasiblity of solving the posted problem using the given dataset and will ensure more accurate analysis results, if proven feasible.

## 2a. What are the assumptions of the distribution of data?

Answer: We are assuming that our data set has normal distribution, meaning that most of the data values are close to the mean.

## 2b. How will we determine the distribution of our data?

Answer: In case of the normal data distribution, mean, mode and median are close. Since all of our columns have numerical data, we can observe these values for each series in the data set as the result of the describe() pandas dataframe method. We can also plot the values for each series using matplotlib library methods with X axis representing individual values of the data and Y axis - the number of rows with those values. Normally distributed data will have bell-shaped plots.

## 3a. How might outliers impact our analisys?

Answer: Outliers often indicate an abnormal condition and/or an error in the process used to collect the data, therefore, they may skew the results of the data analysis and thus are usually ignored in the analysis.

## 3b. How will we test for outliers?

Answer: Since both *admit* and *prestige* variables are categorical and represented by numbers, we will check for rows with values NaN and for rows with values that are no in 0 or 1 for *admit* series and not in 1, 2, 3 or 4 for *prestige* series. For outliers test in other series in the dataset, we can look for rows with deviation that is two or more times (2 is arbitrarily picked) higher than the standard deviation for each series. 

## 4a. What is collinearity?

Answer: Collinearity is a high correlation of the independent variables. This results in inflation of at least one regression coefficient. When collinearity is found, one of the ways to address it is not to include all of the correlated variables into the analysis but include only a subset of variables.

## 4b. How will you test for colinearity?

Answer: Calculate correlation coefficients for each pair of independed variables. If the absolute value of the coefficient is close to 1, the two variables are highly correlated.
