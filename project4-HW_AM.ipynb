Answer: The aim of this project is to determine how the prestige of the applicants' undegraduate school affects their chances of admission to a graduate school.

### Dataset:  Write up a description of your data and any cleaning that was completed

Answer: The data for this project has been generated for the purpose of logistic regression analysis demonstration. 

The data set consists of 4 variables: student's GRE, GPA, prestige of the undegraduate school (categories 1 through 4, with 1 being the most prestigious and 4 - the least) and the admission decision (two categories: 1 - admitted, 0 - not admitted). 

There were 400 observations in the dataset. However, only 397 of those observations contained data for each variable. While imputing missing variables with mean or median values is a common practice, dropping observations with missing variables from the analysis was chosen as the approach for this project, given that incomplete observations were less than 1 percent of all data.

### Methods: Write up the methods used in your analysis

Answer: Dummy binary variables for college prestige 2 through 4 were added to the data set while the prestige column was dropped. The logictic regression model was used to evaludate the odds of admission for students attending colleges, ranked 2 through 4, against the odds of admission for students who attended the highest-ranked colleges. 
We further evaluated the model and the relationship between the prestige of the undegraduate college and the odds of the graduate school admission by creating a fresh dataset with the same, evenly spaced, GRE and GPA values for each prestige category and predicting the odds of admission using our model. 

### Results: Write up your results

Answer: According to the results of the model, the odds of admission increase as the prestige category of the college decreases (college rank gets higher).

### Discussion: Write up your discussion and future steps

Answer: The results of the model confirm the common sense. All 4 tables above confirm the inverse relationship between the prestige of the undergraduate school and the student's changes of admission to graduate school: the higher the prestige category, the lower are the odds of admission.

Looking at table A, we see that, while the average GPA and GRE scores of admitted students are higher than those of unadmitted and that those scores are closer to the average than for the unadmitted students, they do not differ drastically. However, we see the distinct pattern of decreasing the number of admitted and increasing the number of unadmitted students as the prestige category of the undegraduate school gets higher (rank gets lower).

Looking at table B, we see the increase of the inverse coefficient of the admission chances with the increase of the prestige category, as compared to the admission chances for students attending the highest-ranked colleges (lowest prestige category value).

Looking at table C, we see that the odds of admission go down almost 51 percent for students attending # 2-ranked colleges, as compared to odds for students attending the top-ranked colleges.

Looking at table D, with GRE and GPA values being the same (and at the highest possible vaues for both variables), we can clearly see the inverse relationship between the prestige of the college and the probability of the grad school admission: the decrease in the value of the prestige variable (increase in the rank of the college) results in the increase of the grad school admission probability for students with the same GRE and GPA values across all college ranks. The difference in admission probability for students of # 3 and # 4 ranked colleges is less significant than probability difference for students of # 2 and # 3 ranked colleges and for students of # 1 and # 2 ranked colleges. This is also evident in the prestige frequency percentage pattern in table A, prestige coefficients pattern in table B, and the prestige odds ratio in table C. Therefore, the results of the research are consistent throughout.
